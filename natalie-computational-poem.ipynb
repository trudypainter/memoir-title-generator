### TAROT_MEANINGS

```python
req = requests.get("https://raw.githubusercontent.com/dariusk/corpora/master/data/divination/tarot_interpretations.json")

tarot_meanings = []
for entry in req.json()['tarot_interpretations']:
    tarot_meanings.extend(entry['meanings']["light"])
    tarot_meanings.extend(entry['meanings']["shadow"])

print(len(tarot_meanings))
```

### PROVERBS

```python
req = requests.get("https://raw.githubusercontent.com/dariusk/corpora/master/data/words/proverbs.json")

temp_proverbs = []
proverbs = []
for entry in req.json()['proverbs']:
    for key in entry.keys():
        temp_proverbs.extend((entry[key]))

for proverb in temp_proverbs:
    proverbs.append(proverb[:-1])
        
print(len(proverbs))
```
